#Difference of Gaussian





```python
from skimage import data
from skimage.color import rgb2gray
from matplotlib import pyplot as plt
import numpy as np
import skimage.io
from matplotlib.pyplot import imshow, axis
import skimage.color
from skimage.io import imread
import cv2
```

```python
image= data.retina()
image=  rgb2gray(image)
plt.imshow(image, cmap='gray')
plt.title("gray image")
plt.show()
```

```python
img1= cv2.GaussianBlur(image,(31,31),1)
img2=cv2.GaussianBlur(image,(31,31),2)
img3=cv2.GaussianBlur(image,(31,31),3)
img4=cv2.GaussianBlur(image,(31,31),4)
img5=cv2.GaussianBlur(image,(31,31),5)
img6= cv2.GaussianBlur(image,(31,31),6)
img7=cv2.GaussianBlur(image,(31,31),7)

blurred_images = [img1, img2, img3, img4, img5, img6, img7]
sigma_values = [1, 2, 3, 4, 5, 6, 7]

plt.figure(figsize=(15, 20))

for i, img in enumerate(blurred_images):
  plt.subplot(1, 7, i + 1)
  plt.imshow(img, cmap='gray')
  plt.title(f"Sigma = {sigma_values[i]}")
  plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
img10=cv2.GaussianBlur(image,(31,31),10)
```

```python
dog1_7=np.abs(img1 - img7)
dog2_7=np.abs(img2 - img7)
dog3_7=np.abs(img3 - img7)
dog4_7=np.abs(img4 - img7)
dog5_7=np.abs(img5 - img7)
dog6_7=np.abs(img6 - img7)
dog3_4= np.abs(img3 - img4)
dog7_10= np.abs(img7 - img10)
diff = [dog1_7, dog2_7, dog3_7, dog4_7, dog5_7, dog6_7, dog3_4, dog7_10]
values = [1_7, 2_7, 3_7, 4_7, 5_7, 6_7, 3_4, 7_10]

plt.figure(figsize=(12, 6))

for i, img in enumerate(diff):
  plt.subplot(2,4 ,( i + 1))
  plt.imshow(img, cmap='gray')
  plt.title(f"Sigma = {values[i]}")
  plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
threshold= (np.max(dog1_7))*0.2
dog1_7[dog1_7>threshold]=255
dog2_7[dog2_7>threshold]=255
dog3_7[dog3_7>threshold]=255
dog4_7[dog4_7>threshold]=255
dog5_7[dog5_7>threshold]=255
dog6_7[dog6_7>threshold]=255
dog7_10[dog7_10>threshold]=255

th = [dog1_7, dog2_7, dog3_7, dog4_7, dog5_7, dog6_7, dog7_10]
sigma_values = [1_7, 2_7, 3_7, 4_7, 5_7, 6_7, 7_10]

plt.figure(figsize=(12, 8))

for i, img in enumerate(th):
    plt.subplot(3, 3, i + 1)
    plt.imshow(img, cmap='gray')
    plt.title(f"Sigma = {sigma_values[i]}")
    plt.axis('off')

plt.tight_layout()
plt.show()

```

the conclusion :

for the given image retina , gaussian filter of standard deviations 1, 2 ,3, 4 , 7 and 10 are applied then difference of gaussian is determined for different pairs of gaussian filters with low and high standard deviations it is observed that for the pair of std deviation 1 and 7 small as well as large edges are highlighted while for std deviation pair of 4 and 7 only boundary and few small edges are detected , this is because std deviation of 4 has already blurred the small edges , this leads to small difference for small edges and large difference for the large edeges or the boundary of the object .

for difference of gaussian using standard deviation of 3 and 4 the imag representation of dog is almost black this is because blurring of the image is almost the same for two different values of sigma which leads to very low difference between the two images
