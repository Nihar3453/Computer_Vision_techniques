Aim: Identify corners in the image using Harris corner detector

```python
import cv2
import matplotlib.pyplot as plt

image = cv2.imread('/content/window.jfif')

image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
plt.imshow(image_rgb)
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(gray_image, cmap='gray')
plt.title('Grayscale Image')
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
R= cv2.cornerHarris(gray_image, blockSize=2, ksize=5, k=0.04)
plt.subplot(1, 2, 1)
plt.imshow(R)
plt.title('corner harris')
plt.axis('off')
plt.show()
R.shape
```

```python
th= 0.5*R.max()
[rows,cols]= R.shape
for row in range(rows):
  for col in range(cols):
    if R[row,col] > th:
      cv2.circle(image,(col,row),10,(255,0,0),2)


plt.imshow(image)
plt.title('corner')
plt.axis('off')
plt.show()

```
